# Chỉ lấy POI ở VietNam

```python
# Define file paths
viamo_file = 'viamo_full.csv'
new_data_file = 'new_data_final_filled.csv'
filtered_output_file = 'data.csv'  # file mới chứa dữ liệu đã lọc

# Read viamo_full.csv to get all valid IDs
viamo_df = pd.read_csv(viamo_file)
valid_ids = set(viamo_df['id'].unique())

print(f'Number of valid IDs in viamo_full.csv: {len(valid_ids)}')

# Read new_data_final_filled.csv
new_data_df = pd.read_csv(new_data_file)
print(f'Original number of rows in new_data_final_filled.csv: {len(new_data_df)}')

# Filter to keep only rows with IDs in viamo_full.csv
filtered_df = new_data_df[new_data_df['id'].isin(valid_ids)]
print(f'Number of rows after filtering: {len(filtered_df)}')
print(f'Number of rows removed: {len(new_data_df) - len(filtered_df)}')

# Save the filtered dataframe to a new file
filtered_df.to_csv(filtered_output_file, index=False)
print(f'\nFiltered data has been saved to {filtered_output_file} successfully!')
```
